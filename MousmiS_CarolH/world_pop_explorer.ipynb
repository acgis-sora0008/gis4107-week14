```python
# Display only the country names

df = pd.read_csv(out_csv)

df = df[df['Country / Area'].notna()]

# Display only the country names
Country_Names = df['Country / Area']
print(Country_Names)
```

```python
# Display only the Country and July 2023 populations

df = pd.read_csv(out_csv)

df = df[df['Country / Area'].notna()]

selected_columns = ['Country / Area', 'Population (1 July 2023)']
subset_df = df[selected_columns]

print(subset_df)
```

```python
# Display names of the top five most populous countries


df = pd.read_csv(out_csv)


df = df[df['Country / Area'].notna()]
df = df[df['Country / Area'] != 'World']


sorted_df = df.sort_values(by='Population (1 July 2023)', ascending=False)


top_five_countries = sorted_df.head(5)['Country / Area']
print(top_five_countries)

```

```python
# Display the population for the country Réunion (France)

df = pd.read_csv(out_csv)


reunion_population01 = df.loc[df['Country / Area'] == 'Réunion (France)',  'Population (1 July 2022)']
reunion_population02 = df.loc[df['Country / Area'] == 'Réunion (France)',  'Population (1 July 2023)']


print(f" Population of Réunion (France) in 2022: {reunion_population01.values[0]} and Population of Réunion (France) in 2023: {reunion_population02.values[0]}")
```

```python
# Display the name and populations (2023) for all countries with populations less than 50 million
# and sort from highest to lowest population

df = pd.read_csv(out_csv)

filtered_df = df[df['Population (1 July 2023)'] < 50000000]


sorted_df = filtered_df.sort_values(by='Population (1 July 2023)', ascending=False)


result_df = sorted_df[['Country / Area', 'Population (1 July 2023)']]
print(result_df)
```

```python
# Display the list of Continents
df = pd.read_csv(out_csv)

df = df[df['UN Continental Region[4]'].notna()]

# Display only the continents names
Continent_Names = df['UN Continental Region[4]']
print(Continent_Names)
```

```python
# Display the total population for each continent

df = pd.read_csv(out_csv)

continent_population01 = df.groupby('UN Continental Region[4]')['Population (1 July 2022)'].sum()
continent_population02 = df.groupby('UN Continental Region[4]')['Population (1 July 2023)'].sum()

# Display the total population for each continent
print(f" Total Population in 2022: {continent_population01.values[0]} and Total Population in 2023: {continent_population02.values[0]}")
```

```python
# Display the number of countries per continent (HINT:  .groupby)


df = pd.read_csv(out_csv)


countries_per_continent = df.groupby('UN Continental Region[4]')['Country / Area'].count()


print(countries_per_continent)
```
